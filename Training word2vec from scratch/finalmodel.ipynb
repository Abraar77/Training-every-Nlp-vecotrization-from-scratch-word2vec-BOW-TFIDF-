```python
import pandas as pd
```

```python
messages = pd.read_csv('spamCollection.csv', sep='\t',names=["label", "message"])
messages.head()
```

```python
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
lam=WordNetLemmatizer()

import nltk
nltk.download('wordnet')

nltk.download('stopwords')
```

```python
corpus=[]
for i in range(0,len(messages)):
     review=re.sub("[^a-zA-Z0-9]"," ", messages["message"][i])
     review=review.lower()
     review=review.split()
     review=[lam.lemmatize(word) for word in review if word not in set(stopwords.words("english"))]
     review=" ".join(review)
     corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=2500, binary=True, ngram_range=(1,2))
X=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
y=pd.get_dummies(messages["label"])
Y=y.iloc[:,1].values
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.20, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()

params_grid={
    "n_estimators":[50,100,150],
    "min_samples_split":[2,3,5],
    "max_depth":[3,4,6],
    "min_samples_leaf":[1,2,3]


}


grid=GridSearchCV(estimator=model, param_grid=params_grid,scoring="accuracy", cv=5)
grid.fit(X_train,Y_train)
```

```python
best_params=grid.best_params_
best_params
```

```python
best_model=grid.best_estimator_
```

```python
from sklearn.metrics import accuracy_score

y_pred = best_model.predict(X_test)
accuracy = accuracy_score(Y_test, y_pred)
accuracy
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize

import nltk
nltk.download('punkt_tab')
```

```python
words=[]

for word in range(len(corpus)):
    words.append(word_tokenize(corpus[word]))
```

```python
words
```

```python
from gensim.models import Word2Vec


# Train Word2Vec model
model = Word2Vec(words, vector_size=100, window=5, min_count=2, workers=4, sg=1)

# # Save the model for future use
# model.save("word2vec.model")
```

```python
try:
    similarity_cat_dog = model.wv.similarity('cat', 'dog')
    print(f"Similarity between 'cat' and 'dog': {similarity_cat_dog:.4f}")
except KeyError as e:
    print(f"KeyError: {e}")

# Most similar words to 'cat'
try:
    similar_words_cat = model.wv.most_similar('cat', topn=5)
    print("Most similar words to 'cat':", similar_words_cat)
except KeyError as e:
    print(f"KeyError: {e}")

# Perform an analogy task
try:
    analogy_result = model.wv.most_similar(positive=['king', 'woman'], negative=['man'], topn=1)
    print("Analogy result for 'king - man + woman':", analogy_result)
except KeyError as e:
    print(f"KeyError: {e}")
```

```python
import numpy as np

def avg_word2vec(docs):
  valid_words=[model.wv[word] for word in docs if word in model.wv.index_to_key]
  if not valid_words:
    return np.zeros(model.wv.vector_size)
  return np.mean(valid_words,axis=0)
```
